```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = np.array([2.25, 1.7, 1.53, 1.44, 1.34, 1.26])
l = np.array([0.06, 0.08, 0.1, 0.12, 0.15, 0.17])
```

```python

fig = plt.figure(figsize=(10, 8))
plt.rcParams['text.usetex'] = True
plt.rc('xtick', labelsize=20)
plt.rc('ytick', labelsize=20)

x = np.linspace(0.05, 0.18, 100)
k = 0.24
y = np.sqrt(k/x)
plt.errorbar(l, f, xerr=0.005, yerr=0.1, fmt=",", label="Experiment")
plt.plot(x, y, color="red", label=r"Theory for $\mu_d \approx 0.012$")

plt.legend(fontsize=20)
plt.ylabel("Oscillation frequency $f_s$ / Hz", fontsize=22)
plt.xlabel("Distance $L$ / m", fontsize=22)
plt.savefig("plot.png", dpi=200)
```
